```python
# Ranking of Mutual Funds just number  based on esg_score column

df_mutual.groupby('fund_symbol')['esg_score'].mean().sort_values(ascending=False). head(20)

 
```

```python
# how many unique values in esg_score column
df_mutual.esg_score.nunique()
```

```python
# define a new coloumn called esg_ranking based on esg_score column and sort values in descending order---method avg

df_mutual['esg_ranking']=df_mutual['esg_score'].rank(ascending=False)

# show me the output of esg_ranking column and esg_score column and fund_symbol column in descending order-- 
df_mutual[['esg_ranking','esg_score','fund_symbol']].sort_values(by='esg_ranking',ascending=True).head(30)
```

```python
# define a new coloumn called esg_ranking based on esg_score column and sort values in descending order---method min

df_mutual['esg_ranking']=df_mutual['esg_score'].rank(ascending=False, method='min')

# show me the output of esg_ranking column and esg_score column and fund_symbol column in descending order-- 
df_mutual[['esg_ranking','esg_score','fund_symbol']].sort_values(by='esg_ranking',ascending=True).head(30)
```
